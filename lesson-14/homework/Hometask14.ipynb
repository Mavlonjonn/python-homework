```python
#Q3 JSON modification 
import json

def load_books():
    try:
        with open('books.json', 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        return []

def save_books(books):
    with open('books.json', 'w') as file:
        json.dump(books, file, indent=4)

def add_book(books):
    title = input("Enter book title: ")
    author = input("Enter author: ")
    year = input("Enter publication year: ")
    book = {"title": title, "author": author, "year": year}
    books.append(book)
    save_books(books)
    print("Book added successfully!")

def update_book(books):
    title = input("Enter the title of the book to update: ")
    for book in books:
        if book["title"].lower() == title.lower():
            new_author = input(f"Enter new author (current: {book['author']}): ") or book['author']
            new_year = input(f"Enter new year (current: {book['year']}): ") or book['year']
            book["author"] = new_author
            book["year"] = new_year
            save_books(books)
            print("Book updated successfully!")
            return
    print("Book not found!")

def delete_book(books):
    title = input("Enter the title of the book to delete: ")
    for i, book in enumerate(books):
        if book["title"].lower() == title.lower():
            del books[i]
            save_books(books)
            print("Book deleted successfully!")
            return
    print("Book not found!")

def show_books(books):
    if not books:
        print("No books in the list!")
    else:
        print("\nList of Books:")
        for i, book in enumerate(books, 1):
            print(f"{i}. Title: {book['title']}, Author: {book['author']}, Year: {book['year']}")

def main():
    books = load_books()
    while True:
        print("\n1. Add book")
        print("2. Update book")
        print("3. Delete book")
        print("4. Show list of books")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            add_book(books)
        elif choice == "2":
            update_book(books)
        elif choice == "3":
            delete_book(books)
        elif choice == "4":
            show_books(books)
        elif choice == "5":
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```

```python
#Q4 Movie recommendation 

import requests
import random

def get_movies_by_genre(genre):
    api_key = ""  #server is not working temporarily 
    url = f"http://www.omdbapi.com/?s={genre}&type=movie&apikey={api_key}"
    
    try:
        response = requests.get(url)
        data = response.json()
        
        if data.get("Response") == "True":
            return data["Search"]
        else:
            print(f"Error: {data.get('Error', 'No movies found for this genre')}")
            return []
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

def recommend_movie(genre):
    movies = get_movies_by_genre(genre)
    if movies:
        movie = random.choice(movies)
        api_key = ""  #server is not working temporarily 
        url = f"http://www.omdbapi.com/?i={movie['imdbID']}&apikey={api_key}"
        response = requests.get(url)
        movie_details = response.json()
        
        if movie_details.get("Response") == "True":
            print(f"\nRecommended Movie: {movie_details['Title']}")
            print(f"Year: {movie_details['Year']}")
            print(f"Genre: {movie_details['Genre']}")
            print(f"Plot: {movie_details['Plot']}")
        else:
            print("Could not fetch movie details.")
    else:
        print("No movies found to recommend.")

def main():
    while True:
        genre = input("Enter a movie genre (e.g., Action, Comedy, Drama) or 'exit' to quit: ").strip()
        if genre.lower() == "exit":
            print("Exiting...")
            break
        if genre:
            recommend_movie(genre)
        else:
            print("Please enter a valid genre.")

if __name__ == "__main__":
    main()
```
