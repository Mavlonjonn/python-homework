```python
#ToDo class
class Task:
    def __init__(self, title, description, duedate, status='Incomplete'):
        self.title = title 
        self.description = description 
        self.duedate = duedate 
        self.status = status
    def __str__(self):
        return f'Title: {self.title}\n Description: {self.description}\n Due date: {self.duedate}\n Status: {self.status}'
    

class TodoList:
    def __init__(self):
        self.tasks = []

    def newtask(self, ntask):
        self.tasks.append(ntask) 
        print(f'Task {ntask.title} has been added to the list')

    def status(self, title):
        for x in self.tasks:
            if x.title.lower() == title.lower():
                x.status = "Completed"
                print(f'The task {title} marked as Completed')
                return 
        print('Task {title} is not found')
    
    def alltasks(self):
        if self.tasks == []:
            print("There is no task in the list")
            return
        print("Here are all tasks: ")
        for index, ntask in enumerate(self.tasks, 1):
            print(f'Task {index}:')
            print(ntask)

    def incomplete(self):
        itasks = [task for task in self.tasks if task.status == "Incomplete"]
        if itasks == []:
            print("There is no Incomplete task")
            return 
        print("Incomplete tasks: ")
        for index, ntask in enumerate(itasks, 1):
            print(f"Task {index}:\n {ntask}")
    
from IPython.display import clear_output
import time 

def todo():
    print("Welcome to To-do-list App menu: ")
    print("1. Add new task to the list")
    print("2. List all tasks")
    print("3. Mark task as completed")
    print("4. List only incomplete tasks")
    print("5. Exit") 
    
def main():
    todolist = TodoList()

    while True:
        todo()
        choice = input("Enter a number between 1 and 5: ")

        if choice == '1':
            title = input("Enter a title of the task: ")
            description = input('Enter a description of the task: ')
            duedate = input("Enter due date(format:YYY-MM-DD): ")
            newtask = Task(title,description, duedate)
            todolist.newtask(newtask) 
        elif choice == '2':
            todolist.alltasks()
        elif choice == '3':
            utitle = input("Enter title to be marked as Completed: ")
            todolist.status(utitle)
        elif choice == '4':
            todolist.incomplete()
        elif choice == '5':
            print("Exiting the program")
            break 
        else: 
            print("Invalid choice! Please enter number between 1 and 5")
        clear_output(wait=True)
        time.sleep(3)
main()
```

```python
#Blog class
class Post:
    def __init__(self, title, content, author):
        self.title = title 
        self.content = content 
        self.author = author 

    def __str__(self):
        return f'Title: {self.title}\n Content: {self.content}\n Author: {self.author}' 

class Blog:
    def __init__(self):
        self.posts = [] 

    def addpost(self, npost):
        self.posts.append(npost)
        print(f'The post {npost.title} has been added to the list')
    
    def listallposts(self):
        if self.posts == []:
            print("There is no post in the blog")
            return 
        print("Here are the posts: ")
        for index, npost in enumerate(self.posts, 1):
            print(f' Post {index}\n {npost.title}')
    
    def listbyauthor(self, author):
        autpost = [post for post in self.posts if post.author.lower() == author.lower()]
        if autpost == []:
            print("There is no post by {author}.")
            return 
        print("Post(s) by {author}: ")
        for index, npost in enumerate(autpost, 1):
            print(f'The post {index}:\n {npost}')
    
    def deletepost(self, dpost):
        lenpost = len(self.posts)
        self.posts = [post for post in self.posts if post.title != dpost.title]
        if len(self.posts) > lenpost:
            print(f"The post with title {dpost} has been removed successfully")
        else: 
            print(f'There is no post with title {dpost}')
    
    def editpost(self, oldtitle, newtitle, newcontent, newauthor):
        for x in self.posts:
            if x.title == oldtitle:
                x.title = newtitle 
                x.content = newcontent 
                x.author = newauthor 
                print(f'Post {x} has been updated')
                return
        print(f'There is no post with {oldtitle}')
    
    def latestpost(self):
        if self.posts == []:
            print("There are no posts in the list")
            return
        print(f'The latest post: \n {self.posts[-1]}') 
              
from IPython.display import clear_output
import time  

def blogpost():
    print("Welcome to BlogPosts portal: ")
    print('1. Add a new post')
    print('2. List all posts')
    print('3. Display by author')
    print('4. Delete a post')
    print('5. Edit a post')
    print('6. Latest post')
    print('7. Exit the portal')

def main():
    blog = Blog()

    while True:
        blogpost()
        choice = input("Enter a number between 1 and 7")

        if choice == '1':
            title = input("Enter title of the post: ")
            content = input("Enter content of the post: ")
            author = input("Enter author of the post: ") 
            Newpost = Post(title, content, author)
            blog.addpost(Newpost)
        
        elif choice == '2':
            blog.listallposts()
        
        elif choice == '3':
            a = input("Enter author of the post: ")
            blog.listbyauthor(a)
        
        elif choice == '4':
            d = input("Enter the title of the post to be deleted: ")
            blog.deletepost(d) 
        
        elif choice == '5':
            otitle = input("Enter old title of the post to edit: ")
            for x in blog.title:
                if blog.title != otitle:
                    print("There is no post with title {otitle}")
                    return
                else:
                    ntitle = input("Enter new title of the post: ")
                    ncontent = input("Enter new content of the post: ")
                    nauthor = input("Enter new author of the post: ")  
            for x in blog.posts:
                if x.title.lower() == otitle.lower():
                    x.title = ntitle
                    x.content = ncontent
                    x.author = nauthor 
                    print(f'Post {otitle} has been updated')
                    return 
            print("There is no post with title {oldtitle}")
        
        elif choice == '6':
            if blog.posts == []:
                print('There is no post in the list')
                return 
            blog.latestpost()
               
        elif choice == '7':
            print("Exiting the portal")
            break 
        else: 
            print("Invalid number! Enter number in this range: 1-7")
        clear_output(wait=True)
        time.sleep(3)
main()
```
